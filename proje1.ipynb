Bu projede bir öğrenci not sistemi oluşturacağız.

DERS: Algebra

**NOT ARALIĞI:**
*   A - (91-100)
*   B - (80-91)
*   C - (60-81)
*   D - (49-61)
*   F - (0-49) 



Öğrencinin aldığı puanın harf karşılığını gösteren fonksyon yazıyoruz.

```python
def harf(x):
  if x>0 and x<49:
    return "F"
  elif x>49 and x<61:
    return "D"
  elif x>60 and x<81:
    return "C"
  elif x>80 and x<91:
    return "B"
  elif x>90 and x<=100:
    return "A"
  else:
    return "error"
```

Öğrencinin geçib geçmediğini kontrol eden fonksyon yazıyoruz.

```python
def gecerlilik(g):
  if harf(g)=='A' or harf(g)=='B' or harf(g)=='C' or harf(g)=='D':
    return 'gecti'
  elif harf(g)=='F':
    return 'kaldi'
```

Öğrenci Bilgilerini (Ad, Soyad, Okul No, sınav puanı) girebileceğimiz ve bu bilgilerin tutulabileceği bir sistem oluşturuyoruz.

```python
butun_ogrenciler=[]

devam=1

while devam!=0:
  ad=input("Öğrenci adı: ")
  soyad=input("Öğrenci soyadı: ")
  okul_n=input("Okul numarası: ")
  puan=int(input("Sınav puanı: "))
  ogrenci={}
  ogrenci['ad']=ad
  ogrenci['soyad']=soyad
  ogrenci['okul_numarasi']=okul_n
  ogrenci['puan']=puan
  ogrenci['harfli']=harf(puan)
  ogrenci['gecerlilik']=gecerlilik(puan)
  butun_ogrenciler.append(ogrenci) 
  devam=int(input('''Devam:
1:evet 0:hayir
'''))
```

Notları girilen öğrencilerden dersi geçenleri ve geçmeyenleri gösteren bir Dataframe oluşturuyoruz.

```python
import pandas as pd

data = pd.DataFrame(data=butun_ogrenciler)

print(data)
```

Oluşturulan Dataframe’i Excel tablosuna dönüştürünüz

```python
data.to_csv('data.csv')
```
